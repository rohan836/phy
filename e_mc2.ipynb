# Mass-Energy Equivalence ($E = mc^2$)

Albert Einstein's famous equation describes how mass and energy are two forms of the same thing. 

$$\mathbf{E = mc^2}$$

### Variable Meanings
- **$E$ (Energy)**: Measured in Joules (J).
- **$m$ (Mass)**: Measured in kilograms (kg).
- **$c$ (Speed of light)**: $\approx 299,792,458\text{ m/s}$ (speed of light in a vacuum).
- **$c^2$**: $\approx 8.98755 \times 10^{16}\text{ m}^2/\text{s}^2$ (the enormous conversion factor).

### Physical Implications
- **Stellar Fusion**: The Sun fuses hydrogen into helium, converting mass into sunlight.
- **Nuclear Power**: Reactors fission Uranium atoms, converting nuclear mass into electrical energy.
- **Matter Creation**: High energy collisions in particle accelerators can produce new physical matter from energy.

```python
# Core Constants & Utility Functions
C = 299_792_458  # Speed of light in m/s
TNT_PER_KG = 4.184e6  # Joules per kg of TNT equivalent

def energy_from_mass(mass_kg: float) -> float:
    """Calculate energy equivalent (Joules) of a given mass (kg)."""
    return mass_kg * C ** 2

def mass_from_energy(energy_j: float) -> float:
    """Calculate mass equivalent (kg) of a given energy (Joules)."""
    return energy_j / C ** 2
```

## Examples of Objects at Rest

Run the cell below to compute the energy equivalent of various physical objects:

```python
EXAMPLES = [
    ('Paperclip', 0.001),
    ('Apple', 0.2),
    ('Human (70 kg)', 70.0),
    ('Earth', 5.972e24),
]

print(f"{'Object':<16} | {'Mass (kg)':<12} | {'Energy Equivalence (Joules)':<28}")
print('-' * 64)
for name, mass in EXAMPLES:
    e = energy_from_mass(mass)
    print(f'{name:<16} | {mass:<12.3e} | {e:<28.3e} J')
```

## Real-World Problems & Calculations

### Problem 1: Hiroshima Bomb ("Little Boy")
The atomic bomb released approximately **$63\text{ TJ}$** ($6.3 \times 10^{13}\text{ J}$) of energy. What was the mass converted to energy?

```python
E_hiroshima = 6.3e13  # Joules
m_hiroshima = mass_from_energy(E_hiroshima)

print(f'Energy released: {E_hiroshima:.1e} J')
print(f'Mass converted:  {m_hiroshima:.6e} kg ({m_hiroshima * 1000:.2f} grams)')
```

### Problem 2: The Sun's Power Output
The Sun radiates energy at a rate of approximately **$3.846 \times 10^{26}\text{ Watts}$** (Joules per second).
- How much mass does it lose every second?
- How much mass does it lose in a year?

```python
P_sun = 3.846e26  # Watts (J/s)
m_sun_per_second = mass_from_energy(P_sun)
m_sun_per_year = m_sun_per_second * 365.25 * 86400

print(f'Mass lost per second: {m_sun_per_second:.3e} kg/s (about {m_sun_per_second / 1e6:.2f} million metric tons/sec)')
print(f'Mass lost per year:   {m_sun_per_year:.3e} kg/yr')
```

### Problem 3: Energy in 1 Gram of Matter
If **$1\text{ gram}$** of any substance were entirely converted into energy, how much energy would it release? What is this energy in terms of TNT equivalent?

```python
mass_1g = 0.001  # kg
energy_1g = energy_from_mass(mass_1g)
tnt_equiv_kg = energy_1g / TNT_PER_KG

print(f'Energy in 1 gram:   {energy_1g:.3e} Joules')
print(f'TNT equivalent:     {tnt_equiv_kg:.3e} kg ({tnt_equiv_kg / 1e6:.2f} kilotons of TNT)')
```

### Problem 4: Powering a Large City
A large metropolitan area uses approximately **$5 \times 10^{12}\text{ Joules}$** of energy daily. What mass of matter must be fully converted into energy to support it for one day?

```python
E_city_daily = 5.0e12  # Joules
m_city_daily = mass_from_energy(E_city_daily)

print(f'Daily city consumption: {E_city_daily:.1e} J')
print(f'Required mass:           {m_city_daily:.6e} kg ({m_city_daily * 1000:.4f} grams)')
```
